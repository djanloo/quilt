```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
df = pd.read_csv("entropy2.csv")
df = df[[col for col in df.columns if "param" in col or "value" in col]]
rename_dict = {c:c.replace("params_", "") for c in df.columns}
df = df.rename(columns=rename_dict)
y = df['value']
df
```

```python
from umap import UMAP
```

```python
from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler().fit_transform(df.drop(columns=['value']).values)
reducer = UMAP(n_neighbors=10)
embedding = reducer.fit_transform(X)
```

```python
fig,ax = plt.subplots()
plt.scatter(*(embedding.T), c=-y)
plt.colorbar()
plt.show()
```

```python
fig,ax = plt.subplots()
x = df['delay_delta']
argso = np.argsort(x)
plt.plot(x[argso], y[argso])
plt.show()
```
